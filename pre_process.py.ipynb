```python
import pandas as pd
from datetime import date, datetime,timedelta
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
```

### Criteria:

- \# of guests
- Time Span
- \# of bedrooms (optional)
- Area (optional)
- \# of beds (optional)
- Reviews (optional)

We start by first loading listings.csv. This contains all the information about each Airbnb listing.

```python
listings = pd.read_csv("data/listings.csv")
listings
```

As seen there are over 100 columns. We'll clean Listings to only get the columns informaiton valueable to us.

```python
listings = listings[['id','accommodates','bedrooms','beds', 'zipcode', 'review_scores_rating']]
listings.head()
```

From here we can filterout listings and collect only the ones that meet the user given criteria

filter_room_details is a function that takes in the criteria given by the user and returns a dataframe with only the rows that meet what we are looking for

```python
# returns a df with rooms that meet room critera

def filter_room_details(df, num_guests, num_bedrooms, num_beds, area, review):
    
    # get rooms greater than max_num_guests
    df_filtered_room = df[(df['accommodates'] ==  num_guests) 
                        & (df['bedrooms']  >=  num_bedrooms)
                        & (df['beds']  >=  num_beds)
                        & (df['review_scores_rating']  >=  review)]
    
    # if area is given get rooms in area
    if area != 0:
        df_filtered_room = df_filtered_room.loc[df_filtered_room['zipcode']  ==  area]

    
    return df_filtered_room
```

Here we can test this function by finding all rooms with: 
- capacity for 8 guests
- located in 92109
- a review score greater than 80

```python
num_guests = 4
zipcode = 92109
minimum_review = 80

df_filter_room_details_test = filter_room_details(listings, num_guests, 0, 0, zipcode, minimum_review)
df_filter_room_details_test
```

Next we need to filter the listings over the users given time span

We start by loading calendar.csv which contains information about each listing over a set of dates

```python
calendar = pd.read_csv("data/calendar.csv")
calendar.head()
```

Once again we will remove the unnecessary columns. This time we are only focused on  the data and the price of the listing.

We will also drop the rows with any missing elements

```python
calendar = calendar[['listing_id','date','price']]
calendar.dropna()
calendar.head()
```

Next we need to convert the price column from a string to an float

dollar_to_number converts a string with a dollar amount into a float

```python
def dollar_to_number(string):
    string = string.replace('$', '')
    string = string.replace(',', '')
    string = string.replace('.', '')
    number = int(string)/100
    return number

#dollar_to_number('$2,050.00')
```

and then we apply the function to our dataframe

```python
calendar['price'] = calendar['price'].apply(dollar_to_number)
```

Finally we will convert the date column to datetime objects

```python
calendar['date'] = pd.to_datetime(calendar['date'])
```

Now we will take the dataframe from filter_room_details and the users given time frame inorder to gather the prices for each listing over the time span given

```python
# returns a df with rooms that meet time critera

def filter_timespan(df_listings, df_calendar, start_date, end_date):
    
    # get a list of days in the timespan
    day_list = []
    day_count = (end_date - start_date).days + 1
    for day in [d for d in (start_date + timedelta(n) for n in range(day_count)) if d <= end_date]:
        day_list.append(day)
    
    # get listing ids from df_listings
    id_list = df_listings['id'].to_list()
    
    #remove ids from df_calendar that are not in id_list
    df_calendar_filtered = df_calendar[df_calendar['listing_id'].isin(id_list)]
    
    #remove dates from df_calendar that are not in day_list
    df_calendar_filtered = df_calendar[df_calendar['date'].isin(day_list)]
    
    # at this point calendar only contains the listings that meet our criteria
    # we can drop the listing_id column
    df_calendar_filtered = df_calendar_filtered.drop('listing_id', axis = 1)
    
    # group items by date and find the average price on that day
    df_calendar_filtered = df_calendar_filtered.groupby(['date']).mean()
    
    #convert and return as series
    series = pd.Series(df_calendar_filtered['price'])
    return series
```

Test to find the average prices between December and January

```python
start_date = date(2019, 12, 1)
end_date = date(2020, 1, 31)

df_timeseries_test = filter_timespan(df_filter_room_details_test, calendar, start_date, end_date)
#df_timeseries_test
```

## Now combined all together

```python
def create_timeseries(df_listings, df_calendar, num_guests, start_date, end_date, min_num_bedrooms, min_num_beds, zipcode, minimum_review):
    
    df_room_details = filter_room_details(df_listings, num_guests, min_num_bedrooms, min_num_beds, zipcode, minimum_review)
    timeseries = filter_timespan(df_room_details, df_calendar, start_date, end_date)
    return timeseries
```
